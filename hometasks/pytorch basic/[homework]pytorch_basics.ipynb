### Общее количество параметров нейронной сети:

на промежуточных слоях нужно научиться задавать количество нейроной.
Будем считать, что на каждом внутреннем слоем 100 нейронов.
Таким образом между 1 и 2: $64*100=6400$, между 2 и 3: $100*100=10000$ параметров, между 3 и 4:$100*10=1000$. 17400 параметров.

Решается задача классификации на 10 классов, предпоследний класс содержит 10 нейронов - меры принадлежности объекта каждому классу.

Пусть $y_0, y_1, \ldots, y_9$ - выходы предпосленего слоя
На последнем слое применяется **softmax** преобразование.

**SOFTMAX**:

Мы хотим иметь вероятности принадлежности к классам, числа не являются вероятностями, сумма не является единица, да и числа на этих нейронных могут быть не положительными.

1. Применяем преобразование чисел в положительные - экспонента: $e^{y_0},e^{y_1}, \ldots, e^{y_n}$
2. Переход к вероятностям - делим на сумму этих чисел каждое число:
$\frac{e^{y_0}}{e^{y_0}+e^{y_1}+ \ldots + e^{y_9}}$

Данные числа положительны и в сумме дают единицу. Если $y_i >y_j$, то это преобразование сохранится. 

SOFTMAX - вероятности классов $0,1,\ldots, 9$

## Обучение нейронных сетей

### Оптимизация функции потерь

Пусть $(x,y)$ - элемент обучающей выборки, $\theta$ - параметр нейросети.

На последнем слое нейронной сети - вероятность классов.
$$ p = (p_0,p_1,\ldots,p_p)^T$$

Чтобы быть увервенным, что нейронная сеть хорошо классифицирует объект необходимо, чтобы вероятность на данном объекте была высокой. Мы можем увеличивать логарифм вероятности. 

Пускай y - правильный ответ на объекте, нам нужно увеличивать вероятность $\log p_y(x) \to \max$,$-\log p_y(x) \to \min$

Такая функция потерь называется **Logloss**, **Категориальная кросс-энтропия**. Мы не можем добиться идеального значения функции потерь. Мы усредняем функцию потери.

Итоговая задача оптимизации:

$$\frac{1}{l} \sum_{(x,y) \in X_{train}} logloss(y,p(x,\theta)) \to \underset{\theta}{\min}$$

p - функция, которая зависит от объекта и от параметров нейронной сети $\theta$. Необходимо минимизировать функцию потерь по параметрам $\theta$.

Функция потерь минимизируется по методу градиентного спуска. Шаг градиентого спуска - изменяем веса в напрвлении противоположному градиенту.

Если задача регрессии, то MSE.

В обучении нейронной сети есть особенность в том,что функция потерь сождержит очень большое количество параметров. Чтобы сделать 1000 шагов градиентного спуска, нужно провести миллиард вычислений, а шагов должно быть гораздо больше.

На помощь приходит прием: **стохастический градиентный спуск**. Объекты для обучения будем подавать поочереди.
На вход нейронной сети подается один объект, считаем функцию потерь и для данного объекта делаем шаг градиентного спуска.

Истина лежит по середине. Мы отдаем на минимазцию нейронной сети **batch** объектами размерами степени двойки, а после вычисляем функцию потерь для batches.

1. Функция потерь
2. Градиент функции потерь по всем нейронной сети
3. Обновление весов с помощью градиентого спуска.

Необходимо научиться считать производные функции потерь по весам нейронной сети. Нужно вспомнить вычисление проивздной композиции:

$f: R^2 \to R^2$, $g: R^2 \to R$, $g(f(x))$
Получится матрица градиентов.

Правило дифференцирования для одномерного случая:
$$f:R \to R, g: R \to R, g(f(x))' = g'(f(x)) \cdot f'(x)$$

Для n-мерного случая:

g(f(x))' является градиентом так как $g(f(x)): R^2 \to R$

Чтобы вычислить градиента ранней сети необходимо знать градиенты более поздних слоев.

### Слой нейронной сети

Чтобы определить слой, необходимо задать:

- **forward perfomance**: $y = f(x;w)$
- **backward perfomance**: $z(x)=z(f(x;w))$

Forward - считает значение слоя нейронной сети на входе этого слоя.
Backward - считает значение градиентов на входе нейронной сети, если имеет значение градиентов на входе в следующем слое сети.

Значение: Слева направа. Градиенты: справа налево.


AutoGrad - автоматическое вычисление градиентов.
TensorFlow, Mxnet, Theano, PyTorch.
Все вычисления можно производить на лету -> более легкий способ оценок.

# 2. Нахождение косинусной близости

Вам даны две матрицы A и B. Необходимо посчитать косинусную близость между строчками матрицы A и столбцами матрицы B. Ответ - матрица чисел, где номер строки - номер строки из матрицы А, а номер столбца - номер столбца из В, от которых бралась косинусная близость.

Напомним, что косинусная близость двух векторов - косинус угла между ними. В n-мерном пространстве косинус угла между веткорами удобнее всего через скалярное произведение:
$$\cos(angle(x, y)) = \frac{x \cdot y}{\left\|x\right\| \left\|y\right\|}$$

(Наша операция очень похожа на умножение матриц)

```python
def get_cos_sim(A, B):
    """
        A, B - torch float tensors
    """
    """
        My function
    """
    A_split = torch.split(A, 1, dim =0)
    B_split = torch.split(B, 1, dim =1)
    t = tuple(map(torch.norm,B_split))
    normed_B = tuple(B_split[i]/t[i] for i in range(0,len(t)))
    o = torch.cat(normed_B)
    normed_B = torch.transpose(torch.transpose(o,0,1).view(5,4),0,1)
    t = tuple(map(torch.norm,A_split))
    normed_A = torch.cat(tuple(A_split[i]/t[i] for i in range(0,len(t))))
    return torch.mean(torch.mm(normed_A, normed_B))

def get_cos_dist(A, B):
    """
        A, B - torch float tensors
    """
    normed_A = A / torch.norm(A, p=2, dim=1, keepdim=True)
    normed_B = B / torch.norm(B, p=2, dim=0)
    return normed_A @ normed_B

def get_cos_sim(A, B):
    """
        A, B - torch float tensors
    """
    return A / A.norm(dim=1)[:,None] @ B / B.norm(dim=0)

def get_cos_sim(A, B):
    a = A.div(torch.norm(A, dim=1, keepdim=True))
    b = B.div(torch.norm(B, dim=0))
    return a.mm(b)

def get_cos_dist(A, B):
    """
        A, B - torch float tensors
    """
    return (A @ B) / (A.norm(dim=1, keepdim=True) @ B.norm(dim=0, keepdim=True))

def get_cos_sim(A, B):
    """
        A, B - torch float tensors
    """
    norm_a = torch.norm(A, dim=1)
    norm_b = torch.norm(B, dim=0)
    multy_AB = torch.mm(A, B)
    norm_ab = torch.mm(norm_a.view(4,1), norm_b.view(1,5))
    
    return multy_AB / norm_ab

def get_cos_sim(A, B):
    div_A = A / A.norm(dim=1).view(A.shape[1], -1)
    div_B = B / B.norm(dim=0)
    return div_A.mm(div_B)

def get_cos_sim(A, B):
    """
        A, B - torch float tensors
    """
    normsA=((A**2).sum(1)**(0.5)).view(-1,1)
    normsB=((B**2).sum(0)**(0.5)).view(1,-1)
    return     A.mm(B)/normsA/normsB
```

# 3. Линейная регрессия

Раньше мы самостоятельно считали производные, чтобы находить веса линейной регрессии с помощью градиентного спуска. Теперь нам нужно использовать для этого PyTorch и его autograd. 

**Важно**: на самом деле .backward не обновляет содержимое матриц с производными (some_tensor.grad), а прибавляет к ним только что посчитаные значения проивзодных. Это значит, что вызвав .backward дважды, вы получите удвоенную производную. Так как мы обновляем веса в цикле и много раз вызываем .backward, то очень быстро мы получим мусор в some_tensor.grad, если не будем его каждый раз обнулять. Таким образом, в конц итериации после использования производных обнулите значения в матрице производных для всех нужных Вам переменных. Делается это вот так 
> some\_tensor.grad.data.zero_()

```python
class LinearRegression:
    def get_loss(self, preds, y):
        """
            @param preds: предсказания модели
            @param y: истиные значения
            @return mse: значение MSE на переданных данных
        """
        # возьмите средний квадрат ошибки по всем выходным переменным
        # то есть сумму квадратов ошибки надо поделить на количество_элементов * количество_таргетов
        e = preds - y 
        return torch.sum(e**2) / e.numel()#YOUR_CODE
    
    def init_weights(self, input_size, output_size):
        """
            Инициализирует параметры модели
            W - матрица размерности (input_size, output_size)
            инициализируется рандомными числами из
            uniform распределения (torch.rand())
            b - вектор размерности (1, output_size)
            инициализируется нулями
        """
        torch.manual_seed(0) #необходимо для воспроизводимости результатов
        self.W = torch.rand(input_size, output_size).type('torch.DoubleTensor').requires_grad_(True)
        self.b = torch.zeros(1, output_size).type('torch.DoubleTensor').requires_grad_(True)

    def fit(self, X, y, num_epochs=1000, lr=0.001):
        """
            Обучение модели линейной регрессии методом градиентного спуска
            @param X: размерности (num_samples, input_shape)
            @param y: размерности (num_samples, output_shape)
            @param num_epochs: количество итераций градиентного спуска
            @param lr: шаг градиентного спуска
            @return metrics: вектор значений MSE на каждом шаге градиентного
            спуска.
        """
        # сделайте вычисления градиентов c помощью Pytorch и обновите веса
        # осторожнее, оберните вычитание градиента в 
        # with torch.no_grad():
        #    some code
        # иначе во время прибавления градиента к переменной создастся очень много нод в дереве операций
        # и ваши модели в будущем будут падать от нехватки памяти
        self.init_weights(X.shape[1], y.shape[1])
        metrics = []
        for _ in range(num_epochs):
            preds = self.predict(X)
            loss = self.get_loss(preds,y)
            loss.backward()
            with torch.no_grad():
                self.W -= self.W.grad * lr
                self.b -= self.b.grad * lr
                self.W.grad.zero_()
                self.b.grad.zero_()
            metrics.append(self.get_loss(preds, y).data)
            
            
        return metrics

    def predict(self, X):
        """
            Думаю, тут все понятно. Сделайте свои предсказания :)
        """
        return X @ self.W + self.b #YOUR_CODE
```

```python
X, Y = datasets.make_regression(n_targets=3, n_features=2, noise=10, random_state=42)
X_train, X_test, Y_train, Y_test = train_test_split(X, Y)
X = torch.from_numpy(X)
Y = torch.from_numpy(Y)
model = LinearRegression()
mse = model.fit(X, Y)
res = torch.stack(mse).mean()
print("{:.3f}".format(res))
print("{:.3f}".format(np.mean(mse)))
```
