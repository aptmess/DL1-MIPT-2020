# Наиболее часто встречающиеся обозначения

## Случайная величина и характеристики

- $\xi, \eta,\zeta$  - случайные величины
- $D\xi , \sigma^2$ - дисперсия случайной величины
- $F_\xi(x), F(x), G(x) $  - функция распределения
- $f_\xi(x), f(x)$ - плотность распределения 
- $E\xi$ - мат. ожидание случайной величины
- $\sigma$ - стандартное отклонение
- $\theta$- неизвестный параметр или характеристика распределения

## Выборка и характеристики

- $X_{[n]} = (X_1,\ldots,X_n)$ - выборка
- $X_i, Y_i, Z_i$ - элементы выборки
- $D^*$ - выборочная дисперсия
- $S^2, s^2$ - несмещенная оценка дисперсии
- $\bar{x}$ - выборочное среднее
- $x_{med}^*, {med}^*$ - выборочная медиана

## Оценки

- $\hat{\theta}, \tilde{\theta}$ - оценки неизвестного параметра
- $S^-, S^+$ - границы доверительного интервала
- $\gamma$ - доверительная вероятность

## Проверка гипотез

- $H_0$ - нулевая гипотеза
- $H_1$ - альтернативная гипотеза
- $\alpha$ - уровень значимости
- $\beta$ - вероятность ошибки 2-го рода
- $\gamma$ - мощность критерия
- $Z$ - статистика критерия
- $S, V_k$ - критическая область
- $\varphi$ - критическая функция
